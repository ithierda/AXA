## Import the packages

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import gdown
```

## Collect the data for the transport

```python
duration_roundtrip_PAR_SIN = 13
#https://www.google.com/search?q=dur%C3%A9e+vol+paris+singapour&ei=E9dsZKuyOdOqkdUP24OpgA0&oq=dur%C3%A9e+vol+par&gs_lcp=Cgxnd3Mtd2l6LXNlcnAQAxgHMgcIABCKBRBDMgcIABCKBRBDMgcIABCKBRBDMgcIABCKBRBDMgcIABCKBRBDMgcIABCKBRBDMgcIABCKBRBDMgcIABCKBRBDMgcIABCKBRBDMgcIABCKBRBDOgoIABBHENYEELADSgQIQRgAUMsDWLsXYJYzaAFwAXgAgAE-iAHbAZIBATSYAQCgAQHIAQjAAQE&sclient=gws-wiz-serp
#in hours


carbonfootprint_PAR_SIN = 628.22
#kg
#https://borisruf.github.io/carbon-footprint-modeling-tool/index.html?id=scenario-c520e4d7-a190-44c4-abaa-8fa7b1c513c8



carbonfootprint_per_day = carbonfootprint_PAR_SIN/duration_roundtrip_PAR_SIN * 24
carbonfootprint_in_35_days = carbonfootprint_per_day*35

#in kilotonnes
carbonfootprint_in_35_days = carbonfootprint_in_35_days/1000

#keep only 3 decimals
carbonfootprint_in_35_days = round(carbonfootprint_in_35_days, 3)

```

## Collect the data for the habitation

```python
dico = {}
dico = dict()
dico['GPT-3 - 590M queries to GPT-3 in january 2023'] = 42.42
dico['Transport - Carbon footprint of a 35-day non-stop kerosene flight'] = carbonfootprint_in_35_days
dico['Habitation - Monthly carbon footprint of population of Lyon and its agglomeration'] = 40.46

```

```python
fig, ax = plt.subplots()

ax.bar(dico.keys(),dico.values())

ax.set_ylabel('Number of kilotonnes')


for i in dico:
    plt.text( i , dico[i] + 1, str(dico[i]), ha='center', va='center')

plt.xticks(rotation=90)
ax.set_title("Comparison of January's CO2 consumption of GPT-3")
```
