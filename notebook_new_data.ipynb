```python
import pandas as pd
```

```python
df = pd.read_csv("donneÃÅes_azure.csv", sep=";")
```

```python
df
```

```python
df = df.drop(df.columns[-1], axis=1)
df = df.drop(df.index[-2:], axis=0)
```

```python
df
```

```python
# Cr√©er une nouvelle DataFrame pour le fichier transform√©
transformed_df = pd.DataFrame(columns=["Date - Year", "Date - Quarter", "Azure Service", "Scope", "Carbon emissions (tCO2e)"])

transformed_df = transformed_df.assign(**{'Date - Year': 2022})
transformed_df = transformed_df.assign(**{'Date - Quarter': "Q1"})


# Parcourir chaque ligne du DataFrame d'origine
for index, row in df.iterrows():
    
    # R√©cup√©rer le nom de l'OPCO
    azure_service = row["OPCO"]
    
    # R√©cup√©rer les noms des colonnes "Scope 1 (tCO2e)", "Scope 2 MB (tCO2e)", etc.
    scope_columns = df.columns[1:-1]
    
    # Parcourir les colonnes "Scope" et leurs valeurs
    for scope_column in scope_columns:
        scope_value = row[scope_column]
        
        # Cr√©er une nouvelle ligne pour chaque "Scope" avec les valeurs appropri√©es
        new_row = {
            "Date - Year": 2022,
            "Date - Quarter": "Q1",
            "Azure Service": azure_service,
            "Scope": scope_column,
            "Carbon emissions (tCO2e)": scope_value.replace(",", "."),  # Remplacer la virgule par un point pour les nombres d√©cimaux
        }

        transformed_df = pd.concat([transformed_df, pd.DataFrame([new_row])], ignore_index=True)


# Sauvegarder le fichier transform√© en CSV
#transformed_df.to_csv("chemin/vers/fichier_transforme.csv", index=False, sep="\t")
```
